```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Exercise 1

- Generate a Gaussian data sample of size $n=300$ with mean 2 and variance 2.
- Plot the histogram of the data.
- Compute the sample mean and variance.
- Let $n$ vary from 5 to 500 (keep the same mean and variance values). Plot the sample mean and variance vs. $n$.

```python
mu = 2
sigma = np.sqrt(2)

data = np.random.normal(mu, sigma, 300)

plt.hist(data)
plt.show()
```

```python
# mean
np.mean(data)
```

```python
# variance
np.var(data)
```

```python
ngrid = list(range(5, 500))
mean_list = []
var_list = []
for n in ngrid:
    data_n = np.random.normal(mu, sigma, n)
    mean_list.append(np.mean(data_n))
    var_list.append(np.var(data_n))
```

```python
plt.plot(ngrid, mean_list)
plt.title('Sample Mean')
plt.show()
plt.plot(ngrid, var_list)
plt.title('Sample Variance')
plt.show()
# This is a graphical illustration of the law of large numbers.
```

## Exercise 2

- Generate $N=10000$ data samples of $n=100$ data points from the distribution $\mathcal U([0,1])$ (uniform).

The mean and variance of $\mathcal U([0,1])$ are respectively $\mu = 1/2$ and $\sigma = 1/12$.

- Let $\bar X^j$ be the sample mean of the $j$th data sample ($j \in \{1,\dots,N\}$) and $Z_j = \frac{\bar X^j - \mu}{\sigma / \sqrt n}$. Plot the histogram of the sample $\{Z_1, \dots, Z_N\}$.

```python
n = 100
N = 10000

data_samples = np.zeros(shape=(n, N))

for j in range(N):
    data_samples[:,j] = np.random.uniform(0, 1, n)
```

```python
mu = 1/2
sigma = 1 / np.sqrt(12)

sample_means = np.mean(data_samples, 0)
sample_z = (sample_means - mu) / (sigma / np.sqrt(n))

plt.hist(sample_z, 50)
plt.show()
# This is a graphical illustration of the central limit theorem.
```

### Exercise 3

- Generate a Gaussian data sample of size 100 with mean $\begin{pmatrix} 2 \\ 1 \end{pmatrix}$ and variance $\begin{pmatrix} 2 & 1 \\ 1 & 2 \end{pmatrix}$.
- Generate a Gaussian data sample of size 100 with mean $\begin{pmatrix} -2 \\ 1 \end{pmatrix}$ and variance $\begin{pmatrix} 2 & -1 \\ -1 & 2 \end{pmatrix}$.
- Plot these two data samples on the same plot with different colors.

```python
data1 = np.random.multivariate_normal([2, 1], [[2, 1], [1, 2]], 300)
data2 = np.random.multivariate_normal([-2, 1], [[2, -1], [-1, 2]], 300)

plt.scatter(data1[:,0], data1[:,1])
plt.scatter(data2[:,0], data2[:,1])
plt.show()
```

### Exercise 5

- Generate a data sample $\mathbf X$ of size $n=100$ from the distribution $\mathcal U([-5,5])$.
- Generate a data sample $\mathbf e$ of size $n$ from the distribution $\mathcal N(0,9)$.
- Compute the vector $\mathbf Y = -1 + 2 \mathbf X + \mathbf e$.
- Plot $\mathbf Y$ vs. $\mathbf X$.
- Add the line of equation $y = -1 + 2x$ to the plot.

```python
n = 100

X = np.random.uniform(-5, 5, n)
e = np.random.normal(0, 3, n)
Y = -1 + 2 * X + e
```

```python
x = np.array(list(range(-5, 6)))
y = -1 + 2 * x
```

```python
plt.scatter(X, Y)
plt.plot(x, y, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Exercise 6

In this exercise, only the function $\verb|numpy.random.normal|$ is allowed for random variable generation.

Let $X \sim \mathcal N(0,1)$ and $Y_1, \dots, Y_n$ be an i.i.d. sample from $\mathcal N(0,1)$.

Then $\sum_{i=1}^n Y_i^2 \sim \chi^2_n$ and $\frac{X}{\sqrt{\sum_{i=1}^n Y_i^2}} \sim t_n$.

- Simulate a data sample of size 500 from the t-distribution with 3 degrees of freedom.
- Plot the histogram of the data.

```python
ndata = 500

data = np.zeros(shape=ndata)

for i in range(ndata):
    x = np.random.normal(0, 1)
    Y = np.random.normal(0, 1, 3)
    data[i] = x / np.sqrt(np.sum(Y**2))
```

```python
plt.hist(data, 30)
plt.show()
```
